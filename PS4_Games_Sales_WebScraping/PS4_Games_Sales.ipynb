```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_best-selling_PlayStation_4_video_games'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
soup
```

```python
soup.find_all('table')[1]
```

```python
sale_table = soup.find('table', class_ = 'wikitable plainrowheaders sortable')
sale_table
```

```python
table_titles = soup.find_all('th')[1:7]
table_titles
```

```python
sale_table_titles = [title.text.strip() for title in table_titles]
print(sale_table_titles)
```

```python
df = pd.DataFrame(columns = sale_table_titles)
```

```python
df
```

```python
soup.find_all('tr')
```

```python
for tag in soup.find_all('sup', {'class': 'reference'}):
    tag.decompose()
```

```python
column_data = soup.find_all('tr')
```

```python
for row in column_data[8:46]:
    row_data = row.find_all('td')
    split_row_data = [data.text.strip() for data in row_data]
    print(split_row_data)

    length = len(df)
    df.loc[length] = split_row_data

```

```python
df
```

```python
df.to_csv('PS4_Games_Sales.csv')
```
