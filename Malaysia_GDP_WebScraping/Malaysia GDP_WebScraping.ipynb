```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/Economy_of_Malaysia'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
soup
```

```python
soup_table = soup.find_all('table')[1]
```

```python
table = soup.find('table', class_ = 'wikitable')
table
```

```python
table_titles = soup_table.find_all('th')[0:9]
table_titles
```

```python
table_titles = [title.text.strip() for title in table_titles]
print(table_titles)
```

```python
df =  pd.DataFrame(columns = table_titles)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data:
  print(row.find_all('td'))
```

```python
for row in column_data:
  row_data = row.find_all('td')
  individual_row_data = [data.text.strip() for data in row_data]
  print(individual_row_data)
```

```python
for row in column_data[1:]:
  row_data = row.find_all('td')
  individual_row_data = [data.text.strip() for data in row_data]

  length = len(df)
  df.loc[length] = individual_row_data
```

```python
df
```

```python
df.to_csv('Malaysia GDP.csv')
```
